# Introduction

Airbnb (https://www.airbnb.co.uk/) is a website for short-term home and apartment rentals. Although it provides hosts with general guidance, it's still hard for the hosts to determine the best price to rent out their houses/rooms. 

The aim of this project is to predict rental price using a data-driven approach.

For this project, location info will be introduced to the model including the property's proximity to certain venues. This will allow the model to put an implicit price on things such as living close to a park or in a supermarket.
